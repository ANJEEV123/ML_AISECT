```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
# Example: drop non-numeric + target column
X = dataset.drop(columns=['classification', 'id'])   # features
y = dataset['classification'].map({'ckd':1, 'notckd':0})   # target
```

```python
# Step 2: Fill/encode missing values (simple fill for demo)
X = X.fillna(0)
```

```python
# Impute missing values in y
y = y.fillna(y.mode()[0])
```

```python
# Step 3: Use Logistic Regression as estimator
model = LogisticRegression(max_iter=10000)
```

```python
# Step 4: Apply RFE to select top 5 features
rfe = RFE(model, n_features_to_select=5)
fit = rfe.fit(X, y)
```

```python
# Step 5: Show selected features
selected_features = X.columns[fit.support_]
print(" Top 5 selected features using RFE:")
print(selected_features)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

import numpy as np
```

```python
# Prepare data using selected features
X_selected = X[selected_features]
```

```python
# Define algorithms to compare
# Step 1: Define algorithms properly as a dictionary
algorithms = {
    'Logistic Regression': LogisticRegression(solver='liblinear'),
    'Random Forest': RandomForestClassifier(random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'SVM': SVC(kernel='rbf', random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42)
}
```

```python
# Step 2: Evaluate using cross-validation
results = {}
for name, algo in algorithms.items():
    cv_results = cross_val_score(algo, X_selected, y, cv=5, scoring='accuracy')
    results[name] = cv_results.mean()
    print(f"{name}: {cv_results.mean():.4f}")
```

```python
# Print the performance of each algorithm
print("Algorithm Performance (Mean Cross-Validation Accuracy):")
for name, accuracy in results.items():
    print(f"{name}: {accuracy:.4f}")

# Find the best performing algorithm
best_algo_name = max(results, key=results.get)
print(f"\nBest Performing Algorithm: {best_algo_name}")
```
