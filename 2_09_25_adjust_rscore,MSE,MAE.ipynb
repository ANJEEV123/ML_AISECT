```python
# Drop ID and Name
dataset = dataset.drop(columns=["Employee_ID", "Name"])
```

```python
dataset
```

```python
# Define X and y
X = dataset[['Age', 'Gender', 'Department', 'Job_Title', 'Experience_Years',
       'Education_Level', 'Location']]
y = dataset["Salary"]
```

```python
X
```

```python
y
```

```python
# 4. One-hot encode categorical variables (gives 0s and 1s)
x= pd.get_dummies(X, drop_first=True)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 5. Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# 6. Train Linear Regression model
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# 7. Predictions
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error
```

```python
# 8. R² Score
r2 = r2_score(y_test, y_pred)

# 9. Adjusted R² Score
n = x_test.shape[0]   # number of observations
p = x_test.shape[1]   # number of predictors
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
```

```python

print("R² Score:", r2)
print("Adjusted R² Score:", adj_r2)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# 4. Random Forest Regressor
# ===============================
rf_reg = RandomForestRegressor(n_estimators=200, random_state=0)
rf_reg.fit(x_train, y_train)
```

```python
y_pred_rf = rf_reg.predict(x_test)

r2_rf = r2_score(y_test, y_pred_rf)
adj_r2_rf = 1 - (1 - r2_rf) * (n - 1) / (n - p - 1)
mse_rf = mean_squared_error(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
```

```python
print("\n===== Random Forest Regression =====")
print("R² Score:", r2_rf)
print("Adjusted R² Score:", adj_r2_rf)
print("Mean Squared Error:", mse_rf)
print("Mean Absolute Error:", mae_rf)
```
