```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
# Load dataset
dataset = pd.read_csv("/content/housing dataset for desicion tree regression.csv")
```

```python
# Convert date to datetime, then numeric
dataset['date'] = pd.to_datetime(dataset['date'])
dataset['date'] = dataset['date'].astype('int64') // 10**9  # seconds since epoch
```

```python

# Label encode categorical columns
for col in ['street', 'city', 'statezip', 'country']:
    le = LabelEncoder()
    dataset[col] = le.fit_transform(dataset[col])
```

```python
# Select features and target
x = dataset[['date', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
             'floors', 'waterfront', 'view', 'condition', 'sqft_above',
             'sqft_basement', 'yr_built', 'yr_renovated', 'street',
             'city', 'statezip', 'country']]
y = dataset[['price']]
```

```python
# Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# ===== Tuned Decision Tree =====
reg = DecisionTreeRegressor(max_depth=10, min_samples_split=10, random_state=42)
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
dt_r2 = r2_score(y_test, y_pred)
```

```python
# ===== Random Forest =====
rf_reg = RandomForestRegressor(n_estimators=200, max_depth=15, random_state=42)
rf_reg.fit(x_train, y_train)
```

```python
rf_pred = rf_reg.predict(x_test)
rf_r2 = r2_score(y_test, rf_pred)
```

```python
print(f"Tuned Decision Tree R² Score: {dt_r2:.4f}")
print(f"Random Forest R² Score: {rf_r2:.4f}")
```
