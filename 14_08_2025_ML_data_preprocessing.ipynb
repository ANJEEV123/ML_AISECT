```python
#import pandas to run dataset
import pandas as pd
```

```python
# load the makaan_dataset
dataset=pd.read_csv("/content/Makaan_data_700pages.csv")
```

```python
# show the dataset information
dataset.info()
```

```python
# remove the column unnamed
dataset = dataset.drop('Unnamed: 0', axis=1)
```

```python
#remove the column facing_direction
dataset=dataset.drop('Facing_direction',axis=1)
```

```python
#remove the column seller_name
dataset=dataset.drop('Seller_name',axis=1)
```

```python
#add the two columns size, size_unit naming bhk
dataset['BHK'] = dataset['Size'].astype(str) + dataset['Size_unit']
```

```python
#remove the column Size ,Size_unit
dataset = dataset.drop(['Size',	'Size_unit'], axis=1)
```

```python
# Replace any repeated 'BHK' with just one 'BHK'
dataset['BHK'] = dataset['BHK'].str.replace(r'(BHK)+$', 'BHK', regex=True)
```

```python
#Replace any repeated 'apartment' with just one 'apartment'
dataset['Property_type'] = dataset['Property_type'].str.replace(r'(Apartment)+$', 'Apartment', regex=True)
```

```python
# change the crore words to float
dataset['Security_deposit'] = (dataset['Security_deposit'].astype(str).str.strip().str.replace(',', '', regex=False).str.replace(r'^[Nn]o$', '0', regex=True).str.replace(r'(\d+(\.\d+)?)\s*[Ll]', lambda m: str(float(m.group(1)) * 100000), regex=True).str.replace(r'(\d+(\.\d+)?)\s*[Cc]r', lambda m: str(float(m.group(1)) * 10000000),  regex=True).astype(float))
```

```python
# change the lakhs to float
dataset['Rent_price'] = (dataset['Rent_price'].astype(str).str.strip().str.replace(',', '', regex=False).str.replace(r'^[Nn]o$', '0', regex=True).str.replace(r'(\d+(\.\d+)?)\s*[Ll]', lambda m: str(float(m.group(1)) * 100000), regex=True).str.replace(r'(\d+(\.\d+)?)\s*[Cc]r', lambda m: str(float(m.group(1)) * 10000000),  regex=True).astype(float))
```

```python
# Replace all NaN values with 0
dataset = dataset.fillna(0)
```

```python
#transforming the status
dataset['Status'] = le.fit_transform(dataset['Status'])
```

```python
#transforming the bhk
dataset['BHK'] = le.fit_transform(dataset['BHK'])
```

```python
##transforming the location
dataset['Location'] = le.fit_transform(dataset['Location'])
```

```python
#transforming the seller_type
dataset['Seller_type'] = le.fit_transform(dataset['Seller_type'])
```

```python
import numpy as np
# Define price bins (adjust ranges as you want)
bins = [0, 10000, 25000, 50000, np.inf]  # 0-10k, 10k-25k, 25k-50k, >50k
labels = ['Low', 'Medium', 'High', 'Premium']
```

```python

# Create classification target
dataset['Rent_class'] = pd.cut(dataset['Rent_price'], bins=bins, labels=labels)
```

```python
# Encode class labels into integers
from sklearn.preprocessing import LabelEncoder
le_target = LabelEncoder()
dataset['Rent_class_encoded'] = le_target.fit_transform(dataset['Rent_class'])
```

```python
# Features
X = dataset[['Property_type', 'Location', 'Seller_type', 'Area_sqft',
             'Status', 'Security_deposit', 'Bathroom', 'BHK']]
```

```python
# Target (encoded class)
y = dataset['Rent_class_encoded']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
dtc = DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
```

```python
y_pred_dtc = dtc.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dtc))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dtc, target_names=labels))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200, random_state=0)
rfc.fit(X_train, y_train)
```

```python
y_pred_rfc = rfc.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rfc))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rfc, target_names=labels))
```
